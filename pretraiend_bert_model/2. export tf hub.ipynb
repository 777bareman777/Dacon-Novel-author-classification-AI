```python
import tensorflow as tf
from official.nlp.bert.export_tfhub import create_bert_model

bert_config = BertConfig.from_json_file('./bert_model/bert_config.json')
model_checkpoint_path = './bert_model/model.ckpt-1000000'
hub_destination = './export_dir'
vocab_file = './bert_model/vocab.txt'

core_model, encoder = create_bert_model(bert_config)
checkpoint = tf.train.Checkpoint(model=encoder, encoder=encoder)
```

```python
checkpoint.restore(model_checkpoint_path)
```

```python
do_lower_case = 'uncased' in vocab_file

core_model.vocab_file = tf.saved_model.Asset(vocab_file)
core_model.do_lower_case = tf.Variable(do_lower_case, trainable=False)
core_model.save(hub_destination, include_optimizer=False, save_format='tf')
```
